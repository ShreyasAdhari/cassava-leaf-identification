```python
def get_rand_bbox(w,h,lam):

  cut_r = np.sqrt(1-lam)

  cut_w = np.int( w*cut_r )
  cut_h = np.int( h*cut_r )

  cx = np.random.randint(w)
  cy = np.random.randint(h)

  x1 = np.clip( (cx - (cut_w // 2) ), 0, w )
  x2 = np.clip( (cx + (cut_w // 2) ), 0, w )

  y1 = np.clip( (cy - (cut_h // 2) ), 0, h )
  y2 = np.clip( (cy + (cut_h // 2) ), 0, h )

  return x1,x2,y1,y2


def cutmix(X,Y,alpha):

  ## BATCH SHAPE B, C, H, W

  indices = torch.randperm(X.shape[0])

  lam = np.clip( np.random.beta(alpha,alpha),0.3,0.4 )
  x1,x2,y1,y2 = get_rand_bbox(X.shape[-2],X.shape[-1],lam)
  new_X = X.clone() 

  new_X[:,:,y1:y2,x1:x2] = X[indices,:,y1:y2,x1:x2]

  del X
  gc.collect()

  lam = 1-( ( (x2-x1)*(y2-y1) )/( new_X.shape[-1]*new_X.shape[-2] ) )

  return (new_X,Y.squeeze(),Y[indices].squeeze(),lam)
```

```python
from albumentations import (
    HorizontalFlip, VerticalFlip, IAAPerspective, ShiftScaleRotate, CLAHE, RandomRotate90,
    Transpose, ShiftScaleRotate, Blur, OpticalDistortion, GridDistortion, HueSaturationValue,
    IAAAdditiveGaussianNoise, GaussNoise, MotionBlur, MedianBlur, IAAPiecewiseAffine, RandomResizedCrop,
    IAASharpen, IAAEmboss, RandomBrightnessContrast, Flip, OneOf, Compose, Normalize, Cutout, CoarseDropout, ShiftScaleRotate, CenterCrop, Resize
)

from albumentations.pytorch import ToTensorV2

def get_train_transforms():
    return Compose([
            RandomResizedCrop(CFG['img_size'], CFG['img_size']),
            Transpose(p=0.5),
            HorizontalFlip(p=0.5),
            VerticalFlip(p=0.5),
            ShiftScaleRotate(p=0.5),
            HueSaturationValue(hue_shift_limit=0.2, sat_shift_limit=0.2, val_shift_limit=0.2, p=0.5),
            RandomBrightnessContrast(brightness_limit=(-0.1,0.1), contrast_limit=(-0.1, 0.1), p=0.5),
            Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], max_pixel_value=255.0, p=1.0),
            CoarseDropout(p=0.5),
            Cutout(p=0.5),
            ToTensorV2(p=1.0),
        ], p=1.)
  
        
def get_valid_transforms():
    return Compose([
            CenterCrop(CFG['img_size'], CFG['img_size'], p=1.),
            Resize(CFG['img_size'], CFG['img_size']),
            Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], max_pixel_value=255.0, p=1.0),
            ToTensorV2(p=1.0),
        ], p=1.)
```

```python
class CassavaDataset(Dataset):
    
    def __init__(self,file,train_path,tfms,test=False):
        
        self.file = file
        self.path = train_path
        self.tfms = tfms
        self.test = test
    
    def __getitem__(self,index):
        
        img = cv2.cvtColor(cv2.imread( os.path.join(self.path,self.file.iloc[index].image_id) ),cv2.COLOR_BGR2RGB)
        label = self.file.iloc[index].label
        
        if self.tfms:
            img = self.tfms(image=img)['image']
        
        if self.test:

          return img, torch.tensor(np.eye(5, dtype='float64')[label])
        
        else : return img,torch.LongTensor([label])
    
    def __len__(self):
        return self.file.shape[0]
```

```python
def lincomb(a,b,e):
  return (a*e) + ((1-e)*b)

def reduce_loss(loss,reduction="mean"):
  
  if reduction == "mean": return loss.mean()
  elif reduction == "sum" : return loss.sum()
  else : return loss

class LabelSmoothingCrossEntropy(nn.Module):

  def __init__(self,epsilon=0.1,reduction='mean'):
    super().__init__()
    self.epsilon = epsilon
    self.reduction = reduction

  def forward(self,output,target):

    n = output.shape[-1]
    log_preds = F.log_softmax(output,dim=-1)
    log_loss = reduce_loss(-log_preds.sum(dim=-1),self.reduction)
    nll = F.nll_loss(log_preds,target,reduction=self.reduction)

    return lincomb(log_loss/n,nll,self.epsilon)
```

```python
class config:

  lr = 2e-4 
  opt = torch.optim.AdamW
  opt_params = {'weight_decay':1e-6}

  criterion = CutMixCrossEntropyLoss
  criterion_params = {'size_average':True}

  scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts
  scheduler_params = {}
  epoch_scheduler = False
  step_scheduler = True

  epochs = 7
  cutmix = True

  metrics = [torchmetrics.functional.accuracy]
  metrics_names = ['loss','accuracy']
```

```python
class AverageMeter(object):
    """Computes and stores the average and current value"""
    def __init__(self):
        self.reset()

    def reset(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count
```

```python
class TPUFitter():

  def __init__(self,model,config,device,fold=None):

    self.model = model
    self.device = device
    self.config = config
    self.fold = fold



    self.epoch = 0
    

    self.criterion = self.config.criterion(**self.config.criterion_params)

    
    self.opt = self.config.opt( self.model.parameters() , lr=self.config.lr*xm.xrt_world_size(), **self.config.opt_params )

    self.best_acc = 0

    xm.master_print(f'Fitter prepared. Device is {self.device}')
  
  def log(self,x):
    xm.master_print(x)
  
  def fit(self,train_loader,validate_loader):
    
    if self.config.scheduler is not None:
      total_iters = ( len(train_loader) ) * self.config.epochs
      self.scheduler = self.config.scheduler(self.opt,T_0=len(train_loader),T_mult=1, eta_min=1e-6,**self.config.scheduler_params )
      
    
    

    for epoch in range(self.config.epochs):
      

      
      
      self.model.train()
      st = time.time()
      para_loader = pl.ParallelLoader(train_loader,[self.device])
      train_metrics = self.train_one_epoch(para_loader.per_device_loader(self.device))
      xm.master_print("="*100)
      s = '[TRAIN] '+f' Epoch : {epoch} ' +'   '.join(f" {v} : {k:6f} " for k,v in zip(train_metrics,self.config.metrics_names) ) + f' Time : {int(time.time()-st)}'
      self.log(s)
      del para_loader
      gc.collect()
      
      self.model.eval()
      st = time.time()
      para_loader = pl.ParallelLoader(validate_loader,[self.device])
      validate_metrics = self.validate(para_loader.per_device_loader(self.device))
      s = '[VALID] '+f' Epoch : {epoch} ' +'   '.join(f" {v} : {k:6f} " for k,v in zip(validate_metrics,self.config.metrics_names) ) + f' Time : {int(time.time()-st)}'
      self.log(s)
      del para_loader
      gc.collect()

      curr_acc =  xm.mesh_reduce('validation_acc',validate_metrics[1],np.mean)  

      if curr_acc > self.best_acc :

        self.best_acc = curr_acc
        m = f'model_fold_{self.fold}.pth' if self.fold is not None else 'best_model.pth'
        xm.save(self.model.state_dict(),m)


      
      
        
        
      xm.master_print("="*100)
      xm.master_print("\n\n")

    
    
      
  
  def validate(self,validate_loader):

    avg_loss = AverageMeter()
    avg_loss.reset()
    mets = [avg_loss]

    if len(self.config.metrics) > 0:
      for m in self.config.metrics:
        ms = AverageMeter()
        ms.reset()
        mets.append(ms)
      
    

    with torch.no_grad():

      for bi,(x,y) in enumerate(validate_loader):

        x = x.to(self.device)
        
        y = y.to(self.device)
        
        out = self.model(x)
        loss = self.criterion(out,y)

        bs = y.shape[0]

        mets[0].update(loss.detach().item(),bs)
        
        
        if len (mets) > 1 :
          for m,meter in zip ( self.config.metrics,mets[1:] ) :
            sc = m( torch.max(out,dim=-1)[1], torch.max(y,dim=-1)[1]  )
            meter.update(sc.item(),bs)
            del sc
            gc.collect()
        
        del loss,out,x,y
        gc.collect()
    

    return [ m.avg for m in mets ]
  
  def train_one_epoch(self,train_loader):
    
    avg_loss = AverageMeter()
    avg_loss.reset()
    mets = [avg_loss]

    if len(self.config.metrics) > 0 and not self.config.cutmix :
      for m in self.config.metrics:
        ms = AverageMeter()
        ms.reset()
        mets.append(ms)
    

    for i,(x,y) in enumerate(train_loader):

      bs = y.shape[0]

      x = x.to(self.device)
      y = y.to(self.device)
     
      
      out = self.model(x)

      loss = self.criterion(out,y)
        
      


      self.opt.zero_grad()
      loss.backward()

      mets[0].update(loss.detach().item(),bs)
        
        
      if len (mets) > 1 and not self.config.cutmix :
        for m,meter in zip ( self.config.metrics,mets[1:] ) :
          sc = m( torch.max(out,dim=-1)[1], y )
          meter.update(sc.item(),bs)
          del sc
          gc.collect()
      
      xm.optimizer_step(self.opt)
      if ( self.scheduler is not None ) and ( self.config.step_scheduler ):
        self.scheduler.step()

 
  
      
 

      del loss,out,x,y

      
        
    
    if ( self.scheduler is not None )  and ( self.config.epoch_scheduler ) :
        self.scheduler.step()

    return [ m.avg for m in mets]
    
```

```python
def _mp_fn(rank,flags,index):
    
  bs = 16

  device = xm.xla_device()
  model.to(device)

  train_data = CutMix( CassavaDataset(train,'/content/train',train_tfms,test=False), num_class=5, num_mix=2, beta=1.0, prob=0.5 )
  validate_data = CassavaDataset(test,'/content/train',validate_tfms,test=True)

  train_sampler = torch.utils.data.DistributedSampler(train_data,num_replicas=xm.xrt_world_size(),rank=xm.get_ordinal())
  validate_sampler = torch.utils.data.DistributedSampler(validate_data,num_replicas=xm.xrt_world_size(),rank=xm.get_ordinal())

  train_loader = DataLoader(train_data,batch_size=bs,sampler=train_sampler,pin_memory=True,drop_last=True,num_workers=4)
  validate_loader = DataLoader(validate_data,batch_size=bs,sampler=validate_sampler,pin_memory=True,drop_last=False,num_workers=4)

  fitter = TPUFitter(model,config,device,fold=index)

  fitter.fit(train_loader,validate_loader)
```
